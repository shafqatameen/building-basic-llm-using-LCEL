```python
import os 
from dotenv import load_dotenv
load_dotenv()

import openai
openai.api_key=os.getenv("OPENAI_API_KEY")
```

```python
import os 
from dotenv import load_dotenv
load_dotenv()
import langchain_groq
groq_api_key=os.getenv("GROQ_API_KEY")
```

```python
from langchain_openai import ChatOpenAI
from langchain_groq import ChatGroq
model=ChatGroq(model="openai/gpt-oss-120b",groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

messages=[
    SystemMessage(content="Translate the following text into english to urdu"),
    HumanMessage(content="hello, how are you?")

]

result=model.invoke(messages)
result
```

```python
from langchain_core.output_parsers import StrOutputParser
parser=StrOutputParser()
parser.invoke(result)
```

```python
#using LCEL to chain the model and parsec
chain = model | parser
chain.invoke(messages)
```

```python
from langchain_core.prompts import ChatPromptTemplate
generic_tamplet="translate the given into {language}:"
prompt=ChatPromptTemplate.from_messages(
    [("system",generic_tamplet),("user","{text}")]
)
```

```python
result=prompt.invoke({"language":"telugu","text":"hello"})
result
```

```python
result.to_messages()
```

```python
chain = prompt|model | parser 
chain.invoke({"language":"urdu","text":"hello, my dear friend"})
```
